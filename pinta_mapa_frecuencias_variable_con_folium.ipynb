```
box = {'latsup':40.52338262,'latinf':40.35582384, 'lonizd':-3.74096452, 'londch':-3.53640968}

def dentrocajaLAT(lat):
    if lat < box['latsup'] and lat > box['latinf']:
        return True
    else:
        return False

def dentrocajaLON(lon):
    if abs(lon) < abs(box['lonizd']) and abs(lon) > abs(box['londch']):
        return True
    else:
        return False

for clave in box: 
    print(f'{clave}:{box[clave]}')
```

```
import folium
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/corregido_localidades_geolocalizadas.csv', sep = '\t')

df['dentrocajaLAT'] = df['LAT'].apply(dentrocajaLAT)
df['dentrocajaLON'] = df['LON'].apply(dentrocajaLON)

data = df.loc[(df['enmadrid'] == True)]# & 
data = data.loc[(data['dentrocajaLAT'] == False) | (data['dentrocajaLON'] == False)]
data.loc[data['recontado'] > 20, ['recontado']] = 20
data.loc[data['recontado'] == 0, ['recontado']] = 1
print(data['name'])

data.head(50)
data.to_csv('solomadrilenhas.csv')
```

```
this_map = folium.Map(location=[40.520517, -3.680365],
        zoom_start=9.3,tiles='Stamen Toner')
        #prefer_canvas=True)

def escogecolor(lat,lon):
    if lat >= 40.6797 : return '#0bb2db'
    elif lat <= 40.2511: return '#dba00b'
    elif lon >= -3.5433: return '#dba00b'
    elif lon <= -3.9361: return '#30c70a'
    else: return '#bd2409'

def tamanoporsaliencia(saliencia):
    if saliencia < 0.0001: return 2
    elif saliencia > 0.0001 and saliencia < 0.001: return 4
    elif saliencia > 0.001 and saliencia < 0.01: return 8
    elif saliencia > 0.01 and saliencia < 0.1: return 16
    elif saliencia > 0.1: return 32
    else : return 2

def plotDot(point):
    '''input: series that contains a numeric named latitude and a numeric named longitude
    this function creates a CircleMarker and adds it to your this_map'''
    folium.CircleMarker(location=[point.LAT, point.LON],
                        radius=tamanoporsaliencia(point.salience),
                        weight=0,
                        popup = '\n'.join([point['name'],f'cuenta: {point.recontado}',f'saliencia: {round(point.salience,5)}']),
                        fill = True,
                        fill_opacity = 0.5,
                        fill_color = escogecolor(point.LAT, point.LON)).add_to(this_map)    

#use df.apply(,axis=1) to "iterate" through every row in your dataframe
data.apply(plotDot, axis = 1)


#Set the zoom to the maximum possible
#this_map.fit_bounds(this_map.get_bounds())

#Save the map to an HTML file
this_map.save('madridrural_salienciaplot.html')

this_map
```
